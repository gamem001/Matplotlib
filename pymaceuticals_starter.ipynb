```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from sklearn import datasets

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset by merging dataframes
mouse_merge_df = pd.merge(mouse_metadata, study_results, on = 'Mouse ID')
# Display the data table for preview using pd 
mouse_merge_df
```

```python
# Checking the number of mice using unique
mouse_count = len(mouse_merge_df['Mouse ID'].unique())
mouse_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint using.loc.
dup_mouse = mouse_merge_df[mouse_merge_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
#print the ID of the mouse with duplicate entries.
print(dup_mouse.loc[:, 'Mouse ID'].unique())
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_mouse_df = mouse_merge_df[mouse_merge_df["Mouse ID"] == 'g989']
dup_mouse_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data_df = mouse_merge_df.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep=False)
#find all data that is not equal to mouse id g989 (per instructor, drop ALL mice with g989, not just duplicate entries)
clean_data_df = clean_data_df[clean_data_df["Mouse ID"] != 'g989']
clean_data_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# # Capomulin, Ramicane, Infubinol, and Ceftamin

# # Start by getting the last (greatest) timepoint for each mouse
timepoint_df = clean_data_df.groupby('Mouse ID')['Timepoint'].max().reset_index() 
timepoint_df

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tumor_merge = clean_data_df.merge(timepoint_df, on=['Mouse ID', 'Timepoint'])
tumor_merge

#filter out only the drugs I want to see in the dataframe to use as reference for the for loop
filtered_tumor_df = tumor_merge.loc[tumor_merge["Drug Regimen"].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])] 
filtered_tumor_df
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
#find mice treated with capomulin using df with 248 mice
capomulin_mouse = clean_data_df.loc[clean_data_df["Drug Regimen"] == 'Capomulin']
# capomulin_mouse.head()
#locate ALL mice with id s185
mouse_id = clean_data_df[clean_data_df["Mouse ID"] == 's185']
# mouse_id
#create a df of mouse s185 that only identifies timepoint and tumor volume to chart in line
line_data = mouse_id.loc[:, ['Timepoint', 'Tumor Volume (mm3)']]
line_data

# #created line graph using mouse s185
line_data = line_data.plot(x ='Timepoint', y='Tumor Volume (mm3)', kind = 'line', title= "Mouse 's185' Tumor vs Time", color="green")
plt.ylabel("Tumor Volume (mm3)", size= 12)
plt.xlabel("Timepoint", size= 12)
plt.show()
```

```python
#Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
#group ALL mice treated with cap, find mean of their total tumor over the course of study
average_tumor_df = capomulin_mouse.groupby(['Mouse ID']).mean()['Tumor Volume (mm3)']

#create df of mice id by weight, drop duplicate mice
weight_data = capomulin_mouse.loc[:, ['Mouse ID', "Weight (g)"]]
weight_data = weight_data.drop_duplicates()

#merge the two df's to get a new df with id, weight, and tumor average
scatter_data = pd.merge(average_tumor_df, weight_data, on = 'Mouse ID')

#created scatter plot using mouse s185
weight = scatter_data.iloc[:,2]
tumor_volume = scatter_data.iloc[:,1]
plt.scatter(weight,tumor_volume, color= 'red')
plt.ylabel("Tumor Volume (mm3)", size= 12)
plt.xlabel("Mouse Weight (g)", size= 12)
plt.title("Average Tumor Volume vs Mouse Weight", size= 12)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
#created scatter plot again because... just because.
x_values = scatter_data.iloc[:,2]
y_values = scatter_data.iloc[:,1]
correlation = st.pearsonr(weight,tumor_volume)
plt.scatter(weight,tumor_volume, color= "indigo")
plt.title("Average Tumor Volume vs Mouse Weight", size= 12)
plt.xlabel("Weight (g)", size= 12)
plt.ylabel('Average Tumor Volume (mm3)', size= 12)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
## add a linear regression equation - don't forget to import dependencies
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

## taking the equation from the line above, and calculating the points along the line
regress_values = x_values * slope + intercept

plt.scatter(x_values, y_values, color= 'darkblue')
plt.plot(x_values, regress_values, "r-")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept, 2))
plt.annotate(line_eq, (6,10), fontsize = 15, color="red")
plt.title("Mouse weight vs. Avg. Tumor Volume", size= 12)
plt.xlabel("Weight (g)", size= 12)
plt.ylabel('Average Tumor Volume (mm3)', size= 12)
print(f"The r-squared is: {rvalue}")
print(f"The equation of the regression line is: {line_eq}")
plt.show()
```
